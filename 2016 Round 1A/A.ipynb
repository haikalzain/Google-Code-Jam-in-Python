```python
from collections import *
def solve(string):
    letters = Counter(string)
    #numbers = dict(enumerate(['ZERO', 'ONE', 'TWO', 'THREE', 'FOUR', 'FIVE', 'SIX', 'SEVEN', 'EIGHT', 'NINE']))
    numbers = [(2, 'TWO'), (0,'ZERO'),(6,'SIX'),(8,'EIGHT'), (3,'THREE'), (4,'FOUR'),(5,'FIVE'), (7,'SEVEN'), (1,'ONE'), (9,'NINE')]
    result = []
    for val, number in numbers:
        num_count = Counter(number)
        while extract_num(letters, num_count):
            result.append(str(val))
    #print letters
    #assert(sum([b for a,b in letters.iteritems()]) == 0)
    return ''.join(sorted(result))

def extract_num(letters, num_count):
    for l, n in num_count.iteritems():
        if l not in letters or n > letters[l]:
            return False
    
    for l, n in num_count.iteritems():
        letters[l] -= n
    
    return True

infile = open('A-large-practice.in')
#infile = open('test.txt')
outfile = open('out.txt', 'w')
T = int(infile.readline())
for t in range(1, T+1):
    print t
    string = infile.readline()
    outfile.write('Case #{}: {}\n'.format(t, solve(string)))
    
infile.close()
outfile.close()
```
