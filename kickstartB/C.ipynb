```python
def sum_digits(n):
    return sum([int(x) for x in str(n)])
def isprime(n):
    for i in range(2, int(n**0.5 + 1)):
        if n % i == 0:
            return False
    return True
def isgnumber(n):
    return isprime(sum_digits(n))
```

```python
%%time
import primefac
from collections import *
from itertools import combinations
import sys
sys.setrecursionlimit(100000)
def solve(N):
    primes = Counter(list(primefac.primefac(N)))
    factors = [k**v for k,v in primes.iteritems()]
    bits = [False for _ in range(len(factors))]
    mem = {}

    def dp(state, value):
        if value in mem:
            return mem[value]
        if isgnumber(value):
            mem[value] = False
            return False
        new_state = list(state)
        for index, val in enumerate(state):
            if not val:
                new_state[index] = True
                result = dp(tuple(new_state), value / factors[index])
                new_state[index] = False
                if not result:
                    mem[value] = True
                    return True
        mem[value] = False
        return False

    return dp(tuple(bits), N)

infile = open("C-large-practice.in")
outfile = open("out.txt", 'w')
T = int(infile.readline())
for t in range(1, T+1):
    N = int(infile.readline())
    outfile.write("Case #{}: {}\n".format(t, "Laurence" if solve(N) else "Seymour"))

infile.close()
outfile.close()
```
