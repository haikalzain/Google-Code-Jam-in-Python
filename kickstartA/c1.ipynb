```python
def lights(pattern, N): #N is the number of lights, not the index
    complete_sets = N / len(pattern)
    remainder = N % len(pattern)
    return complete_sets * pattern.count('B') + pattern[:remainder].count('B')
    

infile = open('A-large-practice (1).in')
outfile = open('out.txt', 'w')
T = int(infile.readline())
for t in range(1, T+1):
    pattern = infile.readline().strip()
    #print(len(pattern))
    I, J = map(int, infile.readline().split())    
    result = lights(pattern, J) - lights(pattern, I - 1)
    outfile.write("Case #{}: {}\n".format(t, result))

infile.close()
outfile.close()
```
